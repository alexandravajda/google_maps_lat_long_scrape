```python
import pandas as pd
import folium

from IPython.display import IFrame
```

```python
data = pd.read_csv('coastal_towns.csv') 
```

```python
data.head()
```

```python
data.groupby(['size'])['town'].count()
```

```python
data.dropna()
```

```python
data = data[data['lat'].notna()] # found one without lat long
data.head()
```

```python
# Replacing lat long of Sunderland 
data.loc[data['town'].str.contains('Sunderland', case=False),'lat'] = 54.7072602
data.loc[data['town'].str.contains('Sunderland', case=False),'long'] = -2.960353
```

```python
data.loc[data['town'].str.contains('Sunderland', case=False)]
```

```python

seaside_town_large  = data[data['size'] == 'Larger seaside town'] # red
seaside_town_small = data[data['size'] == 'Smaller seaside town']
other_coastal_large  = data[data['size'] == 'Larger other coastal town']
other_coastal_small  = data[data['size'] == 'Smaller other coastal town']

data_map = folium.Map( [52.600,-1.208], tiles='stamenwatercolor' ) #CartoDB positron
   
for lat, long, town in zip(seaside_town_large.lat, seaside_town_large.long, seaside_town_large.town):
    folium.Marker( [lat, long], 
                   icon=folium.CustomIcon( icon_image='https://i.imgur.com/CYx04oC.png', icon_size=(15,15) ), popup=town).add_to(data_map)

for lat, long, town in zip(seaside_town_small.lat, seaside_town_small.long, seaside_town_small.town):
    folium.Marker( [lat, long], 
                   icon=folium.CustomIcon( icon_image='https://imgur.com/Rzs4Zpa.png', icon_size=(10,10) ), popup=town).add_to(data_map)
    
for lat, long, town in zip(other_coastal_large.lat, other_coastal_large.long, other_coastal_large.town):
    folium.Marker( [lat, long], 
                   icon=folium.CustomIcon( icon_image='https://imgur.com/6TWrNOY.png', icon_size=(8,8) ), popup=town).add_to(data_map)
    
for lat, long, town in zip(other_coastal_small.lat, other_coastal_small.long, other_coastal_small.town):
    folium.Marker( [lat, long], 
                   icon=folium.CustomIcon( icon_image='https://imgur.com/C1MXk3r.png', icon_size=(6,6) ), popup=town).add_to(data_map)


data_map.save('coastal_map.html')
IFrame(src='coastal_map.html', width='100%', height=800)
```
